```python
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline

#step_list = ['R2','L3']
#step_list = ['R2', 'R2', 'R2']
#step_list = ['R5', 'L5', 'R5', 'R3']
step_list = ['R3','R1','R4','L4','R3','R1','R1','L3','L5','L5','L3','R1','R4','L2','L1','R3','L3','R2','R1','R1','L5','L2','L1','R2','L4','R1','L2','L4','R2','R2','L2','L4','L3','R1','R4','R3','L1','R1','L5','R4','L2','R185','L2','R4','R49','L3','L4','R5','R1','R1','L1','L1','R2','L1','L4','R4','R5','R4','L3','L5','R1','R71','L1','R1','R186','L5','L2','R5','R4','R1','L5','L2','R3','R2','R5','R5','R4','R1','R4','R2','L1','R4','L1','L4','L5','L4','R4','R5','R1','L2','L4','L1','L5','L3','L5','R2','L5','R4','L4','R3','R3','R1','R4','L1','L2','R2','L1','R4','R2','R2','R5','R2','R5','L1','R1','L4','R5','R4','R2','R4','L5','R3','R2','R5','R3','L3','L5','L4','L3','L2','L2','R3','R2','L1','L1','L5','R1','L3','R3','R4','R5','L3','L5','R1','L3','L5','L5','L2','R1','L3','L1','L3','R4','L1','R3','L2','L2','R3','R3','R4','R4','R1','L4','R1','L5']
#step_list = ['R8','R4','R4','R8']

# starting point
x_list = [0]
y_list = [0]
x = 0
y = 0
facing = 'north'
print(str(x) + ',' + str(y))

# initiate plot
fig = plt.figure()
ax = plt.subplot(111)
ax.set_ylim([-160, -130])   # set the bounds to be 10, 10
ax.set_xlim([0, 40])

#iterate through the step list, calculating a new point for each move
for c in step_list:
    
    new_direction = c[0]
    if (new_direction == 'R'):
        if (facing == 'north'):
            facing = 'east'
        elif (facing == 'east'):
            facing = 'south'
        elif (facing == 'south'):
            facing = 'west'
        elif (facing == 'west'):
            facing = 'north'
    elif (new_direction == 'L'):
        if (facing == 'north'):
            facing = 'west'
        elif (facing == 'west'):
            facing = 'south'
        elif (facing == 'south'):
            facing = 'east'
        elif (facing == 'east'):
            facing = 'north'
    #print(facing)
    
    new_distance = int(c[1:])
    #print(new_distance)
    
    if (facing == 'north'):
        y = y + new_distance
    elif (facing == 'east'):
        x = x + new_distance
    elif (facing == 'south'):
        y = y - new_distance
    elif (facing == 'west'):
        x = x - new_distance
    
    x_list.append(x)
    y_list.append(y)
    print(str(x) + ',' + str(y))
    #print (x_list)
    #print(y_list)

ax.plot(x_list, y_list)
print(x_list)
print(y_list)
final_x = x_list[-1]
final_y = y_list[-1]
print(str(final_x) + ',' + str(final_y))
print('end distance:')
print(abs(final_x) + abs(final_y))


fig.show()
```
