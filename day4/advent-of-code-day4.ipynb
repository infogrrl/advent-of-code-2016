```python
import string
import re
import csv
import collections

#import CSV file and convert to a list
room_list = []
with open('day4.csv', 'r') as csvfile:
    filereader = csv.reader(csvfile)
    for row in filereader:
        row = row[0]
        room_list.append(row)
    print(len(room_list))

q = 0
r = 0
nm = 0
real_room_list = []
#iterate through items, evaluating each one and appending the result to another list
for n in room_list:
    #split the room name into its parts (encrypted name, sector ID, checksum)
    #print(n)
    
    #get encrypted room name as just a set of letters
    component_name = re.split("[0-9]", n)
    component_name = str.replace(component_name[0],"-","")
    #print(component_name)
    
    component_checksum = re.split("\[", n)
    component_checksum = str.replace(component_checksum[1],"]","")
    #print(component_checksum)
    
    p = re.compile('[0-9]+')
    m = p.findall(n)
    component_sectorID = m[0]
    #print(component_sectorID)
    
    #Get the count of each letter
    letters = collections.Counter(component_name)
    #print(letters)
    
    '''
    for letter, count in letters.most_common(5):
        print('%s: %d' % (letter, count))
    print("------")
    print(set(letters))
    print("------")
    '''
    
    letters_dictionary = dict(letters)
    #print(letters_dictionary)
    #Get the top 5 counts, sort same count alphabetically
    d= [v[0] for v in sorted(letters_dictionary.items(), key=lambda kv: (-kv[1], kv[0]))]
    #print(d)
    to_check = d[0] + d[1] + d[2] + d[3] + d[4]
    #print(to_check)
    
    if(to_check==component_checksum):
        q = q + int(component_sectorID)
        #print(q)
        r = r + 1
        real_room_list.append(n)
    else:
        nm = nm + 1

#If this value matches the checksum, get the sector ID and add it to a list

#sum all the sector IDs in the list
```
