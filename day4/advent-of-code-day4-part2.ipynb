```python
import string
import re
import csv
import collections

#import CSV file and convert to a list
room_list = []
with open('day4-part2.csv', 'r') as csvfile:
    filereader = csv.reader(csvfile)
    for row in filereader:
        row = row[0]
        room_list.append(row)
    print("Number of rooms in encrypted list:")
    print(len(room_list))
    print("Here are the components:")
unencrypted_room_list = []
#iterate through items, evaluating each one and appending the result to another list
for n in room_list:
    #split the room name into its parts (encrypted name, sector ID, checksum)
    #print(n)
    e = ""
    #get encrypted room name as just a set of letters
    component_name = re.split("[0-9]", n)
    component_name = str(component_name[0])
    print(component_name)
    
    p = re.compile('[0-9]+')
    m = p.findall(n)
    component_sectorID = m[0]
    print(component_sectorID)
    
    alphabet = ["a","b","c","d","e","f","g","h","i","j","k","l","m","n","o","p","q","r","s","t","u","v","w","x","y","z"]
    #print(alphabet.index("a"))
    code_add = int(component_sectorID)%26
    for letter in component_name:
        #print(letter)
        if (letter=="-"):
            e = e + " "
        else:
            i = alphabet.index(letter)
            i_new = i + code_add
            if (i_new>25):
                i_new = i_new - 26
            decoded_letter = alphabet[i_new]
            e = e + decoded_letter
    print(e)
    print("-----")
        
    
```
